### VAE model with Regressor

```python
class VAE_with_regressor(nn.Module):
    def __init__(self, input_size_dna, input_size_gene, level_2, level_3, latent_dim):
        super(VAE_with_regressor, self).__init__()

        self.input_size_dna = input_size_dna
        self.input_size_gene = input_size_gene
        self.level_2 = level_2
        self.level_3 = level_3
        self.latent_dim = latent_dim
        # Encoder layers
        self.enc_fc1_dna = nn.Sequential(
                        nn.Linear(input_size_dna, level_2),
                        nn.BatchNorm1d(level_2),
                        nn.ReLU())
        
        self.enc_fc1_gene = nn.Sequential(
                        nn.Linear(input_size_gene, level_2),
                        nn.BatchNorm1d(level_2),
                        nn.ReLU())
        
        self.enc_fc2_dna = nn.Sequential(
                        nn.Linear(level_2, level_3),
                        nn.BatchNorm1d(level_3),
                        nn.ReLU())
        
        self.enc_fc2_gene = nn.Sequential(
                        nn.Linear(level_2, level_3),
                        nn.BatchNorm1d(level_3),
                        nn.ReLU())
        
        self.enc_fc3_mean = nn.Sequential(
                    nn.Linear(level_3 + level_3, latent_dim),
                    nn.BatchNorm1d(latent_dim))
        
        self.enc_fc3_log_var = nn.Sequential(
                    nn.Linear(level_3 + level_3, latent_dim),
                    nn.BatchNorm1d(latent_dim))
        
        # Decoder layers

        self.dec_fc3 = nn.Sequential(
                        nn.Linear(latent_dim, level_3+level_3),
                        nn.BatchNorm1d(level_3+level_3),
                        nn.ReLU())
        
        self.dec_fc2_dna = nn.Sequential(
                        nn.Linear(level_3, level_2),
                        nn.BatchNorm1d(level_2),
                        nn.ReLU())
        self.dec_fc2_gene = nn.Sequential(
                        nn.Linear(level_3, level_2),
                        nn.BatchNorm1d(level_2),
                        nn.ReLU())
        
        self.dec_fc1_dna = nn.Sequential(
                    nn.Linear(level_2, input_size_dna),
                    nn.BatchNorm1d(input_size_dna),
                    nn.Sigmoid())
        
        self.dec_fc1_gene = nn.Sequential(
                    nn.Linear(level_2, input_size_gene),
                    nn.BatchNorm1d(input_size_gene),
                    nn.Sigmoid())
        
         # attention layer
        self.attention = MultiheadAttention(embed_dim=latent_dim, num_heads=latent_dim)
        
        # Regression fc layers
        self.r_fc1 = nn.Sequential(
                    nn.Linear(latent_dim, 64),
                    nn.BatchNorm1d(64),
                    nn.ReLU())
        self.r_fc2 = nn.Sequential(
                    nn.Linear(64, 32),
                    nn.BatchNorm1d(32),
                    nn.ReLU())
        
        self.r_fc3 = nn.Sequential(
                    nn.Linear(32, 10),
                    nn.BatchNorm1d(10))
        
    def encode(self, x):
        dna_l2_layer = self.enc_fc1_dna(x[0])
        gene_l2_layer = self.enc_fc1_gene(x[1])
        dna_l3_layer = self.enc_fc2_dna(dna_l2_layer)
        gene_l3_layer = self.enc_fc2_gene(gene_l2_layer)
        l3_layer = torch.cat((dna_l3_layer, gene_l3_layer), 1)

        mu = self.enc_fc3_mean(l3_layer)
        logvar = self.enc_fc3_log_var(l3_layer)
        mu_new = mu.unsqueeze(0).transpose(0, 1)
        attended_mean, _ = self.attention(mu_new,mu_new,mu_new)
        attended_mean = attended_mean.transpose(0, 1).squeeze(0)

        return mu, logvar, attended_mean
    
    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        z = mu + eps * std
        return z
    
    def decode(self, z):
        l3_layer = self.dec_fc3(z)
        dna_l3_layer = l3_layer.narrow(1, 0, self.level_3)
        gene_l3_layer = l3_layer.narrow(1, self.level_3, self.level_3)

        dna_l2_layer = self.dec_fc2_dna(dna_l3_layer)
        gene_l2_layer = self.dec_fc2_gene(gene_l3_layer)

        recon_dna = self.dec_fc1_dna(dna_l2_layer)
        recon_gene = self.dec_fc1_gene(gene_l2_layer)

        return [recon_dna, recon_gene]
    
    def regressor(self, mean):
        level_1_layer = self.r_fc1(mean)
        level_2_layer = self.r_fc2(level_1_layer)
        output_layer = self.r_fc3(level_2_layer)
        return output_layer
    
    def forward(self, x):
        mu, logvar, attended_mean = self.encode(x)
        z = self.reparameterize(mu, logvar)
        x_hat = self.decode(z)
        pred_y = self.regressor(attended_mean)

        return x_hat, mu, logvar, z, pred_y
```

### Define loss functions

```python
def vae_loss_function(x_hat, x, mean, log_var): # recon loss and kld loss
        bce = torch.nn.functional.binary_cross_entropy(x_hat, x, reduction = 'sum')
        kld = -0.5 * torch.sum(1 + log_var - mean.pow(2) - log_var.exp())
        loss = kld + bce
        return loss, kld, bce
```

```python
def rmse_loss_function(y_pred, y):
    loss_fn = torch.nn.MSELoss(reduction='sum')
    loss = torch.sqrt(loss_fn(y_pred, y))
    return loss
```

```python
def mse_loss_function(y_pred, y):
    loss_fn = torch.nn.MSELoss(reduction='sum')
    loss = loss_fn(y_pred, y)
    return loss
```

```python
def mse_loss_function_mean(y_pred, y):
    loss_fn = torch.nn.MSELoss(reduction='mean')
    loss = loss_fn(y_pred, y)
    return loss
```

### Validation Functions

```python
def r_squared(y_true, y_pred):
    y_true_mean = torch.mean(y_true)
    SS_res = torch.sum(torch.square(y_true - y_pred))
    SS_tot = torch.sum(torch.square(y_true - y_true_mean))
    r_squared = 1 - SS_res / (SS_tot + torch.finfo(torch.float32).eps)
    return r_squared
```

### Training Parameters

```python
epochs = 50

lr = 0.001
# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

k = 10
splits = KFold(n_splits=k, shuffle=False)
```

```python
# Initialize the model
input_size_dna = 81037
input_size_gene = 57820 #dimension of gene expressions
level_2 = 2048
level_3 = 1024
latent_dim = 128 # target latent size
model = VAE_with_regressor(input_size_dna, input_size_gene, level_2, level_3, latent_dim).to(device)
optimizer = optim.Adam(model.parameters(), lr=lr)
```

### Train Function

```python
def train (model, dataloader, epoch, optimizer, w_dna_recon_loss, w_gene_recon_loss, w_kl_loss, w_reg_loss):
    model.train()
    train_dna_recon = 0
    train_gene_recon = 0
    train_kl  = 0
    train_reg = 0
    train_total_loss = 0
    batch_num = 0

    for batch_idx, batch in enumerate(dataloader):
        batch_num = batch_num + 1
        # get values
        dna_meth_values, gene_exp_values, drug_resp = batch
        dna_meth_values = dna_meth_values.to(device)
        gene_exp_values = gene_exp_values.to(device)
        drug_resp = drug_resp.to(device)

        dna_plus_gene = [dna_meth_values, gene_exp_values]

        
        # clear gradients
        optimizer.zero_grad()
        # apply model
        x_hat, mu, logvar, z, pred_y = model(dna_plus_gene)

        # losses
        dna_vae_loss, dna_kld, dna_bce = vae_loss_function(x_hat[0], dna_meth_values, mu, logvar)
        gene_vae_loss, gene_kld, gene_bce = vae_loss_function(x_hat[1], gene_exp_values, mu, logvar)
        # gene_bce = mse_loss_function(x_hat[1], gene_exp_values)
        kld = dna_kld
        
        mse_loss = mse_loss_function_mean(pred_y, drug_resp)

        loss = w_dna_recon_loss * dna_bce + w_gene_recon_loss * gene_bce + w_kl_loss * kld + w_reg_loss * mse_loss

        

        # backward
        loss.backward()

        with torch.no_grad():
            train_dna_recon += dna_bce.item()
            train_gene_recon += gene_bce.item()
            train_kl += kld.item()
            train_reg += mse_loss.item()
            train_total_loss += loss.item()

        # update parameters
        optimizer.step()

    train_dna_recon_ave = train_dna_recon/ len(dataloader.sampler)
    train_gene_recon_ave = train_gene_recon/ len(dataloader.sampler)
    train_kl_ave = train_kl/ len(dataloader.sampler)
    train_reg_ave = train_reg/ batch_num
    train_total_ave = train_total_loss/ len(dataloader.sampler)

    print('=====> Epoch {} \n' 
          'Average DNA Recon Loss: {:.3f} \n'
          'Average Gene Recon Loss: {:.3f} \n'
          'Average KL Loss: {:.3f} \n'
          'Average MSE Loss: {:.3f} \n'.format(epoch, train_dna_recon_ave, train_gene_recon_ave, train_kl_ave, train_reg_ave))
    
    return train_dna_recon_ave, train_gene_recon_ave, train_kl_ave, train_reg_ave, train_total_ave

```

### Validation Function

```python
def test(dataloader, model):
    model.eval()
    test_recon_dna = 0
    test_recon_gene = 0
    test_kl = 0
    test_reg = 0
    test_rmse = 0
    test_r2 = 0
    batch_num = 0

    with torch.no_grad():
        pred_store = torch.zeros(1,10).to(device)
        for batch_idx, batch in enumerate(dataloader):
            batch_num = batch_num + 1

            # get data values
            dna_meth_values, gene_exp_values, drug_resp = batch
            dna_meth_values = dna_meth_values.to(device)
            gene_exp_values = gene_exp_values.to(device)
            drug_resp = drug_resp.to(device)

            dna_plus_gene = [dna_meth_values, gene_exp_values]

            # apply model
            x_hat, mu, logvar, z, pred_y = model(dna_plus_gene)

            # loss functions
            dna_vae_loss, dna_kld, dna_bce = vae_loss_function(x_hat[0], dna_meth_values, mu, logvar)
            gene_vae_loss, gene_kld, gene_bce = vae_loss_function(x_hat[1], gene_exp_values, mu, logvar)
            # gene_bce = mse_loss_function(x_hat[1], gene_exp_values)
            kld = dna_kld
            mse_loss = mse_loss_function(pred_y, drug_resp)
            r2_value = r_squared(drug_resp, pred_y)

            test_recon_dna += dna_bce.item()
            test_recon_gene += gene_bce.item()
            test_kl += kld.item()
            test_reg += mse_loss.item()
            test_r2 += r2_value.item()

            # Get relevant values
            pred_store = torch.cat((pred_store, pred_y), 0)

        # ====== export predictions ======
        pred_final = pred_store[1:]
        pred_final_np = pred_final.cpu().numpy()
        pred_final_df = pd.DataFrame(pred_final_np)

        input_path = 'C:/Users/mdzak/Desktop/GitHub/FYP_Zaki/notebooks_new/output/'
        pred_path = input_path + str(latent_dim) + 'D_dna_plus_gene_predictions.tsv'

        pred_final_df.to_csv(pred_path, sep='\t')

        # print loss
        test_recon_dna_ave = test_recon_dna/ len(dataloader.sampler)
        test_recon_gene_ave = test_recon_gene/ len(dataloader.sampler)
        test_kl_ave = test_kl/ len(dataloader.sampler)
        test_reg_ave = test_reg/ (len(dataloader.sampler)*10)
        test_rmse_ave = math.sqrt(test_reg_ave)
        test_r2_ave = test_r2/ batch_num
        
        print('Average DNA Recon Loss: {:.3f} \n'
              'Average Gene Recon Loss: {:.3f} \n'
          'Average KL Loss: {:.3f} \n'
          'Average MSE Loss: {:.3f} \n'
          'Average RMSE Loss: {:.3f} \n'
          'Average R2: {:.3f} \n'.format(test_recon_dna_ave, test_recon_gene_ave, test_kl_ave, test_reg_ave, test_rmse_ave, test_r2_ave))
        
        return test_recon_dna_ave, test_recon_gene_ave, test_kl_ave, test_reg_ave, test_rmse_ave, test_r2_ave
```

### Execute Training and K fold Cross Validation

```python
history_train = {'Train DNA Recon Loss': [], 'Train Gene Recon Loss': [],'Train KL Loss': [], 'Train MSE Loss': [], 'Train Total Loss': []}
history_test = {'Test DNA Recon Loss': [],'Test Gene Recon Loss': [], 'Test KL Loss': [], 'Test MSE Loss': [], 'Test RMSE Loss': [], 'Test R2': []}

for fold, (train_idx,val_idx) in enumerate(splits.split(np.arange(len(full_dataset)))):
    print('Fold {}'.format(fold + 1))
    train_sampler = SubsetRandomSampler(train_idx)
    test_sampler = SubsetRandomSampler(val_idx)
    train_loader = DataLoader(full_dataset, batch_size=batch_size, sampler=train_sampler)
    test_loader = DataLoader(full_dataset, batch_size=batch_size, sampler=test_sampler)

    print("Start Training (Unsupervised Phase)")
    for epoch in range(1, epochs + 1):
        train_dna_recon_ave, train_gene_recon_ave, train_kl_ave, train_reg_ave, train_total_ave = train(model, train_loader, epoch, optimizer, w_dna_recon_loss=0.5, w_gene_recon_loss=0.5, w_kl_loss=1, w_reg_loss=0)
        history_train['Train DNA Recon Loss'].append(train_dna_recon_ave)
        history_train['Train Gene Recon Loss'].append(train_gene_recon_ave)
        history_train['Train KL Loss'].append(train_kl_ave)
        history_train['Train MSE Loss'].append(train_reg_ave)
        history_train['Train Total Loss'].append(train_total_ave)

    print("Start Training (Supervised Phase)")
    for epoch in range(1, epochs + 1):
        train_dna_recon_ave, train_gene_recon_ave, train_kl_ave, train_reg_ave, train_total_ave = train(model, train_loader, epoch, optimizer, w_dna_recon_loss=0, w_gene_recon_loss=0, w_kl_loss=0, w_reg_loss=1)
        history_train['Train DNA Recon Loss'].append(train_dna_recon_ave)
        history_train['Train Gene Recon Loss'].append(train_gene_recon_ave)
        history_train['Train KL Loss'].append(train_kl_ave)
        history_train['Train MSE Loss'].append(train_reg_ave)
        history_train['Train Total Loss'].append(train_total_ave)

    test_recon_dna_ave, test_recon_gene_ave, test_kl_ave, test_reg_ave, test_rmse_ave, test_r2_ave = test(test_loader, model)
    history_test['Test DNA Recon Loss'].append(test_recon_dna_ave)
    history_test['Test Gene Recon Loss'].append(test_recon_gene_ave)
    history_test['Test KL Loss'].append(test_kl_ave)
    history_test['Test MSE Loss'].append(test_reg_ave)
    history_test['Test RMSE Loss'].append(test_rmse_ave)
    history_test['Test R2'].append(test_r2_ave)
```

### Performance of model through 5 Fold CV

```python
train_avg_cv_recon_loss_dna = np.mean(history_train['Train DNA Recon Loss'])
train_avg_cv_recon_loss_gene = np.mean(history_train['Train Gene Recon Loss'])
train_avg_cv_kl_loss = np.mean(history_train['Train KL Loss'])
train_avg_cv_mse_loss = np.mean(history_train['Train MSE Loss'])
train_avg_cv_total_loss = np.mean(history_train['Train Total Loss'])

test_avg_recon_loss_dna =  np.mean(history_test['Test DNA Recon Loss'])
test_avg_recon_loss_gene =  np.mean(history_test['Test Gene Recon Loss'])
test_avg_kl_loss =  np.mean(history_test['Test KL Loss'])
test_avg_mse_loss =  np.mean(history_test['Test MSE Loss'])
test_avg_rmse_loss =  np.mean(history_test['Test RMSE Loss'])
test_avg_r2_loss =  np.mean(history_test['Test R2'])

print('Performance of {} fold cross validation {} dimension:'.format(k, latent_dim))
print("\nAverage Training DNA Recon Loss: {:.4f} \nAverage Training Gene Recon Loss: {:.4f} \n Average Training KL Loss: {:.4f} \n Average Training RMSE Loss: {:.3f} \n Average Training Total Loss: {:.3f} \n".format(train_avg_cv_recon_loss_dna,train_avg_cv_recon_loss_gene, train_avg_cv_kl_loss, train_avg_cv_mse_loss, train_avg_cv_total_loss))
print("\nAverage Testing DNA Recon Loss: {:.4f} \nAverage Testing Gene Recon Loss: {:.4f} \n Average Testing KL Loss: {:.4f} \n Average Testing MSE Loss: {:.3f} \n Average Testing RMSE Loss: {:.3f} \n Average Testing R2 Loss: {:.3f} \n".format(test_avg_recon_loss_dna,test_avg_recon_loss_gene, test_avg_kl_loss, test_avg_mse_loss, test_avg_rmse_loss, test_avg_r2_loss))
```
