```python
import pandas as pd
import numpy as np
```

### Drug Data

```python
drug_resp = pd.read_csv('../results_clean/gdsc_drug_nodash.csv')
drug_resp = drug_resp.drop(columns=['Unnamed: 0'])
drug_resp = drug_resp.sort_values('LN_IC50')
```

```python
drug_resp
```

```python
drug_resp['DRUG_NAME'].unique()
```

```python
chosen_drugs = ['Dactinomycin', 'Dasatinib', 'Daporinad', 'Trametinib',
       'Rapamycin', 'Vinorelbine', 'SN-38', 'Vinblastine', 'Romidepsin',
       'Temsirolimus']
```

```python
drug_resp_chosen = drug_resp[drug_resp['DRUG_NAME'].isin(chosen_drugs)]
```

```python
drug_resp_chosen = drug_resp_chosen.reset_index()
```

```python
drug_resp_chosen.drop(columns=['index'],inplace=True)
```

```python
drug_resp_chosen.to_csv('new_clean_data/drug_resp_top_10.csv')
```

```python
pivot_drug_resp = pd.pivot_table(drug_resp_chosen, values='LN_IC50', index = 'CELL_LINE_NAME', columns= 'DRUG_NAME')
```

```python
pivot_drug_resp[pivot_drug_resp.index == '2313287']
```

```python
drug_resp_chosen[drug_resp_chosen['CELL_LINE_NAME'] == '22rv1']
```

```python
pivot_drug_resp.reset_index(inplace=True)
```

```python
pivot_drug_resp
```

```python
pivot_drug_resp.to_csv('new_clean_data/grouped_drug_resp_na.csv')
```

```python
pivot_drug_resp_no_na = pivot_drug_resp.fillna(pivot_drug_resp.mean())
```

```python
pivot_drug_resp_no_na.to_csv('new_clean_data/grouped_drug_resp_no_na.csv')
```
