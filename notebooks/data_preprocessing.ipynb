### Import Gene Expressions of Cell Lines

### Import RPPA

```python
import pandas as pd
import numpy as np
```

```python
rppa = pd.read_csv('../datasets/CCLE_RPPA_20181003 (1).csv')
```

```python
rppa = rppa.rename(columns={'Unnamed: 0':'CELL_LINE_NAME'})
rppa = rppa.replace('NaN', 0, regex=True)
rppa_drop_na = rppa.dropna(axis=1)
rppa_drop_na
```

```python
rppa_drop_na.to_csv('../datasets_transpose_csv/rppa_transpose.csv')
```

```python
pivot_drug_resp_no_na = pd.read_csv('../notebooks_new/new_clean_data/grouped_drug_resp_no_na.csv')
```

```python
pivot_drug_resp_no_na.drop(columns=['Unnamed: 0'], inplace=True)
pivot_drug_resp_no_na
```

```python
def rem(string):

    if '_' in string:
        return string[:string.index('_')]
    else:
        return string
def lowercase(string):
    new_str = str(string).lower()
    return new_str
```

```python
rppa_drop_na['CELL_LINE_NAME'] = rppa_drop_na['CELL_LINE_NAME'].apply(lowercase)
rppa_drop_na['CELL_LINE_NAME'] = rppa_drop_na['CELL_LINE_NAME'].apply(rem)
```

```python
rppa_drop_na
```

```python
merged_rppa = pd.merge(rppa_drop_na, pivot_drug_resp_no_na, on= 'CELL_LINE_NAME')
```

```python
merged_rppa
```

```python
merged_rppa.to_csv('../notebooks_new/new_clean_data/merged_rppa.csv')
```

```python
dna_meth = pd.read_csv('../notebooks_new/new_clean_data/merged_dna_meth.csv')
dna_meth.drop(columns='Unnamed: 0', inplace=True)
gene_exp = pd.read_csv('../notebooks_new/new_clean_data/merged_gene_exp.csv')
gene_exp.drop(columns='Unnamed: 0', inplace=True)
```

```python
dna_meth_cell_list = dna_meth['CELL_LINE_NAME'].to_list()
gene_exp_cell_list = gene_exp['CELL_LINE_NAME'].to_list()
rppa_cell_list = merged_rppa['CELL_LINE_NAME'].to_list()
```

```python
all_cell_list = list(set(dna_meth_cell_list) & set(gene_exp_cell_list) & set(rppa_cell_list))
```

```python
consistent_dna_meth = dna_meth[dna_meth['CELL_LINE_NAME'].isin(all_cell_list)]
consistent_gene_exp = gene_exp[gene_exp['CELL_LINE_NAME'].isin(all_cell_list)]
consistent_rppa = merged_rppa[merged_rppa['CELL_LINE_NAME'].isin(all_cell_list)]
```

```python
consistent_dna_meth_sorted = consistent_dna_meth.sort_values('CELL_LINE_NAME')
consistent_gene_exp_sorted = consistent_gene_exp.sort_values('CELL_LINE_NAME')
consistent_rppa_sorted = consistent_rppa.sort_values('CELL_LINE_NAME')
```
