```python
import pandas as pd
```

```python
genes = pd.read_csv("../datasets_transpose_csv/genes_transpose.csv")
```

```python
def rem(string):

    if '_' in string:
        return string[:string.index('_')]
    else:
        return string
```

```python
genes
```

```python
genes['CELL_LINE'] = genes['CELL_LINE'].apply(rem)
```

```python
cell_line = genes['CELL_LINE']
```

```python
cell_line
```

```python
def lowercase(string):
    new_str = str(string).lower()
    return new_str
```

```python
genes['CELL_LINE_NAME'] = genes['CELL_LINE'].apply(lowercase)
```

```python
cell_line = genes['CELL_LINE_NAME']
```

```python
cell_line.to_csv('../results_clean/cell_line_name.csv')
```

### Combine with drug response

```python
full_gdsc_drug = pd.read_excel('../datasets/GDSC2_fitted_dose_response_24Jul22 (1).xlsx')
```

```python
full_gdsc_drug
```

```python
gdsc_drug = full_gdsc_drug[['LN_IC50', 'CELL_LINE_NAME','DRUG_NAME']]
```

```python
gdsc_drug
```

```python
def rem_dash(string):

    if '-' in string:
        new_str = str(string).replace('-','')
        return new_str
    else:
        return string
```

```python
gdsc_drug['CELL_LINE_NAME'] = gdsc_drug['CELL_LINE_NAME'].apply(rem_dash)
```

```python
gdsc_drug
```

```python
def lowercase(string):
    new_str = str(string).lower()
    return new_str
```

```python
gdsc_drug['CELL_LINE_NAME'] = gdsc_drug['CELL_LINE_NAME'].apply(lowercase)
```

```python
gdsc_drug
```

```python
gdsc_drug.to_csv("../results_clean/gdsc_drug_nodash.csv")
```
