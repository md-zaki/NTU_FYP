```python
import pandas as pd
```

### Standardize cell_line_names for gdsc drug response data and reduced dimension data

```python
d128 = pd.read_csv('../results_clean/d128_clean.csv')
```

```python
def lowercase(string):
    new_str = str(string).lower()
    return new_str
```

```python
d128 = d128.drop(columns=['Unnamed: 0.1'])
```

```python
d128 = d128.drop(columns=['Unnamed: 0'])
```

```python
d128['CELL_LINE_NAME'] = d128['CELL_LINE_NAME'].apply(lowercase)
```

```python
d128
```

```python
d128.to_csv('../results_clean/d128_clean.csv')
```

```python
d256 = pd.read_csv('../results_clean/d256_clean.csv')
d256 = d256.drop(columns=['Unnamed: 0.1'])
d256 = d256.drop(columns=['Unnamed: 0'])
d256['CELL_LINE_NAME'] = d256['CELL_LINE_NAME'].apply(lowercase)
d256.to_csv('../results_clean/d256_clean.csv')
```

```python
d256
```

```python
d512 = pd.read_csv('../results_clean/d512_clean.csv')
d512 = d512.drop(columns=['Unnamed: 0.1'])
d512 = d512.drop(columns=['Unnamed: 0'])
d512['CELL_LINE_NAME'] = d512['CELL_LINE_NAME'].apply(lowercase)
d512.to_csv('../results_clean/d256_clean.csv')
```

```python
d512
```

```python
d1024 = pd.read_csv('../results_clean/d1024_clean.csv')
d1024 = d1024.drop(columns=['Unnamed: 0.1'])
d1024 = d1024.drop(columns=['Unnamed: 0'])
d1024['CELL_LINE_NAME'] = d1024['CELL_LINE_NAME'].apply(lowercase)
d1024.to_csv('../results_clean/d256_clean.csv')
```

### Combine latent dim and drug response

```python
drug_resp = pd.read_csv('../results_clean/gdsc_drug_nodash.csv')
drug_resp.drop(columns='Unnamed: 0',inplace=True)
```

```python
drug_resp
```

```python
d128
```

```python
d128_drug = pd.merge(d128, drug_resp, on='CELL_LINE_NAME')
d256_drug = pd.merge(d256, drug_resp, on='CELL_LINE_NAME')
d512_drug = pd.merge(d512, drug_resp, on='CELL_LINE_NAME')
d1024_drug = pd.merge(d1024, drug_resp, on='CELL_LINE_NAME')
```

```python
d128_drug.to_csv('../data_for_pred/d128_with_drug.csv')
```
